```python
from sklearn.datasets import load_iris
iris_data = load_iris()
type(iris_data)
```

```python
print(iris_data.keys())
print(iris_data.feature_names)
print(iris_data.target_names)
train = iris_data.data
```

```python
type(train)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(train[:,0],train[:,3],s=150, c=iris_data.target)
plt.xlabel
#plt.show()
```

```python
print(iris_data.data.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
x1,x2, y1,y2 = train_test_split(iris_data.data, iris_data.target, test_size=0.2, random_state=10)

#lgr = LogisticRegression()
lgr = LinearSVC()

lgr.fit(x1,y1)

y_pred = lgr.predict(x2)
print(accuracy_score(y_pred, y2))
```
